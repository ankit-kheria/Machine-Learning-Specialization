```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use("seaborn-dark")
```

```python
data1 = np.random.normal(0, 1, 100_000)
data2 = np.random.normal(0, 0.5, 100_000)
data3 = np.random.normal(0, 2, 100_000)
data4 = np.random.normal(7, 5, 100_000)
```

```python
# How can I sample 1000 points from each of the 4 datasets?
data1_sampled = np.random.choice(data1, 1000)
```

```python
rand_data = np.random.choice(data1, 12)
```

```python
plt.figure(figsize=(14, 8))
plt.hist(data1, bins=30, density=True, alpha=0.5, color='b')
plt.scatter(rand_data, np.zeros((1, 12))+0.01, color='red', s=500, marker="x")
plt.axvline(0, color='#00FA0A', lw=3, alpha=0.7, linestyle="dashed", label="Mean")
plt.fill_between(np.linspace(-1, 1, 100), 0, 0.394, color='red', alpha=0.1, label="Std Dev")
plt.title("Gaussian (Normal) Distribution", fontsize=20)
plt.xlabel("  $x_1$", fontsize=15)
plt.ylabel("$p(x_1)$", fontsize=15)
plt.legend(fontsize=15)
plt.tight_layout()
# plt.show()

# How can I save this figure with some padding?
plt.savefig("gaussian_distribution_sample.png", dpi=300, bbox_inches='tight', pad_inches=0.5, facecolor='w', edgecolor='w')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8), sharey=True)
fig.suptitle('Normal Distribution', fontsize=18)
fig.supylabel("$p(x)$", fontsize=16)
fig.supxlabel("$x$", fontsize=16)

ax[0, 0].set_title('$\mu=0, \sigma=1$')
ax[0, 0].hist(data1, bins=30, density=True, alpha=0.5, color='b')
ax[0, 0].axvline(0, color='red', linestyle='dashed', linewidth=2, label='mean')
ax[0, 0].axvline(1, color='#00FFAD', linestyle='dashed', linewidth=2, label='std')
ax[0, 0].axvline(-1, color='#00FFAD', linestyle='dashed', linewidth=2, label='std')

ax[0, 1].set_title('$\mu=0, \sigma=0.5$')
ax[0, 1].hist(data2, bins=30, density=True, alpha=0.5, color='r')
ax[0, 1].axvline(0, color='red', linestyle='dashed', linewidth=2, label='mean')
ax[0, 1].axvline(0.5, color='#00FFAD', linestyle='dashed', linewidth=2, label='std')
ax[0, 1].axvline(-0.5, color='#00FFAD', linestyle='dashed', linewidth=2)

ax[1, 0].set_title('$\mu=0, \sigma=2$')
ax[1, 0].hist(data3, bins=30, density=True, alpha=0.5, color='g')
ax[1, 0].axvline(0, color='red', linestyle='dashed', linewidth=2, label='mean')
ax[1, 0].axvline(2, color='#00FFAD', linestyle='dashed', linewidth=2, label='std')
ax[1, 0].axvline(-2, color='#00FFAD', linestyle='dashed', linewidth=2)

ax[1, 1].set_title('$\mu=7, \sigma=5$')
ax[1, 1].hist(data4, bins=30, density=True, alpha=0.5, color='y')
ax[1, 1].axvline(7, color='red', linestyle='dashed', linewidth=2, label='mean')
ax[1, 1].axvline(12, color='#00FFAD', linestyle='dashed', linewidth=2, label='std')
ax[1, 1].axvline(2, color='#00FFAD', linestyle='dashed', linewidth=2)

plt.legend(loc='upper right')
plt.tight_layout()
# plt.show()
plt.savefig("./images/normal_distribution-variants.png", dpi=300,  facecolor='w')
```
